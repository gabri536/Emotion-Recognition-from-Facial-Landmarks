The following code follows the https://pyimagesearch.com/2017/04/03/facial-landmarks-dlib-opencv-python/ Tutorial. We firstly import the required Python packages which are not already load:

```python
import dlib
import imutils
from imutils import face_utils
```

Some utility function for convenience:

```python
def rect_to_bb(rect):
	# take a bounding predicted by dlib and convert it
	# to the format (x, y, w, h) as we would normally do
	# with OpenCV
	x = rect.left()
	y = rect.top() #X coordinate from the left and y coordinate from the top.
	w = rect.right() - x
	h = rect.bottom() - y
	# return a tuple of (x, y, w, h)
	return (x, y, w, h)
```

```python
def shape_to_np(shape, dtype="int"):
	# initialize the list of (x, y)-coordinates
	coords = np.zeros((68, 2), dtype=dtype)
	# loop over the 68 facial landmarks and convert them
	# to a 2-tuple of (x, y)-coordinates
	for i in range(0, 68):
		coords[i] = (shape.part(i).x, shape.part(i).y)
	# return the list of (x, y)-coordinates
	return coords
```

Facial landmark extraction:

```python
# initialize dlib's face detector (HOG-based) and then create
# the facial landmark predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(r"C:\Users\Maximilian\Documents\Applied in Paython\shape_predictor_68_face_landmarks.dat")

#Specifying data:
data = train_images_1

#storing the shape data of the images:
shapes = []

for k in range(0, 10):
	# load the input image, resize it, and convert it to grayscale
	#image = cv2.imread(loaded_images[3000])
	image = data[k]
	image = imutils.resize(image, width=500)
	gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

	# detect faces in the grayscale image
	rects = detector(gray, 1)
	# loop over the face detections
	for (i, rect) in enumerate(rects):
		# determine the facial landmarks for the face region, then
		# convert the facial landmark (x, y)-coordinates to a NumPy
		# array
		shape = predictor(gray, rect)
		shape = face_utils.shape_to_np(shape)

		# convert dlib's rectangle to a OpenCV-style bounding box
		# [i.e., (x, y, w, h)], then draw the face bounding box
		(x, y, w, h) = face_utils.rect_to_bb(rect)
		cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
		# show the face number
		cv2.putText(image, "Face #{}".format(i + 1), (x - 10, y - 10),
		cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
		# loop over the (x, y)-coordinates for the facial landmarks
		# and draw them on the image
		for (x, y) in shape:
			cv2.circle(image, (x, y), 1, (0, 0, 255), -1)
			#Collecting shape data of the images:
	shapes.append(shape)

print(type(shapes))
print(len(shapes))
endshapes = np.concatenate(shapes, )
# show the output image with the face detections + facial landmarks
cv2.startWindowThread()
cv2.imshow("Output", image)
cv2.waitKey(20000)
cv2.destroyAllWindows()
# If WaitKey argument = 0 :To stop running the cell press 'Enter' on the image!
```
